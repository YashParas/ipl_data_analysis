```python
import pandas as pd

df = pd.read_csv("matches.csv")
df.head
```

```python
import matplotlib.pyplot as plt

# Count number of matches per season
matches_per_season = df['Season'].value_counts().sort_index()

# Plot it
plt.figure(figsize=(10,5))
plt.bar(matches_per_season.index, matches_per_season.values, color='skyblue')
plt.title('Matches Played per IPL Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(matches_per_season.index)
plt.show()
```

```python
# Count total wins per team
top_teams = df['winner'].value_counts().head(10)

# Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
top_teams.plot(kind='bar', color='orange')
plt.title('Top 10 Teams by Total Wins')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
same_winner = df[df['toss_winner'] == df['winner']]
percentage = (len(same_winner) / len(df)) * 100
print(f"Toss winner also won the match in {percentage:.2f}% of the games.")
```

```python
top_players = df['player_of_match'].value_counts().head(10)
top_players.plot(kind='bar', color='green')
plt.title('Top 10 Player of the Match Winners')
plt.xlabel('Player')
plt.ylabel('Awards')
plt.xticks(rotation=45)
plt.show()
```
